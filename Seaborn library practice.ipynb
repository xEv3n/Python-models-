```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
 
```

```python
print(sns.get_dataset_names())       #Get free dataset from seaborn
```

```python
crash_df = sns.load_dataset("car_crashes")       # loading car_crashes dataset 
crash_df.head()
```

```python
#Distribution Plot 

sns.distplot(crash_df["alcohol"],rug= True, bins= 5)                 #Use Shift + Tab for code reference 
```

```python
#Joint Plot 

sns.jointplot(data= crash_df, x = "speeding",y = "alcohol"  )   # It show two graphical representations in one output 
```

```python
#Kde Plots 

sns.kdeplot(data= crash_df['alcohol'])                       #It is a distribution plot function 
```

```python
#Pair Plot

tips_df = sns.load_dataset("tips")
sns.pairplot(data= tips_df, hue= 'sex', palette="Blues")
```

```python
#Rug Plots 

sns.rugplot(tips_df['tip'])      #where sticks are more they are more frequent 
```

```python
#Styling and Figsize  

sns.set_style("white")
plt.figure(figsize = (8,4))
sns.set_context("talk",font_scale= 1.5)          # It is used to change the size of the visualization
sns.jointplot(data= crash_df, x= "alcohol", y = "speeding", hue= None, kind= "reg")
sns.despine(left=True, right=True, bottom= True , top= True)    # It is used to remove line in the graph not grids 
```

```python
#Categorical Plots
```

```python
#Bar Plots
sns.barplot(data= tips_df , x = "sex", y = "total_bill",estimator= np.var)
sns.set_context("paper",font_scale= 1.5)
```

```python
#Count Plots 
sns.countplot(data= tips_df, x= "sex")      #It is used to calculate number of samples 

   
```

```python
#Box plot
sns.boxplot(data=tips_df, x = "day", y = "total_bill"  , hue= "sex" )    # The middle line between the candle is mean and Open/close is standard deviation and Open/close is also standard deviation

plt.legend(loc = True)
```

```python
#Violin Plot 
sns.violinplot(data=tips_df, x = "day", y = "total_bill"  , hue= "sex" ,split= True)   #It is the combination of box plot and kde plot

```

```python
#Strip Plot
plt.figure(figsize= (8,5))
sns.stripplot(data=tips_df , x = 'day', y = 'total_bill' , jitter= True, hue= "sex")
plt.legend(loc = False)    
```

```python
#Swarm Plot    It is just violin plot but in dotted form 

sns.swarmplot(data=tips_df, x = 'day', y = 'total_bill', color = "Black")
sns.violinplot(data=tips_df, x = "day", y = "total_bill", color = "yellow")
```

```python
#pallete 
plt.figure(figsize= (8,5))
sns.stripplot(data=tips_df , x = 'day', y = 'total_bill' , jitter= True, hue= "sex", palette= "viridis" ) # Go to matplotlib patelle site to check number of options available
plt.legend(loc = False)
```

```python
#Matrix Plot 
```

```python
#Heatmap 
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index= "month", columns= "year", values="passengers")
sns.heatmap(data=flights,cmap= "Blues")
```

```python
# Cluster Map 
iris = sns.load_dataset("iris")
species  = iris.pop("species")
sns.clustermap(iris)
```

```python
#Pair Grids 
iris  = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

```python
#Regression Plot 

plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale= 1.5)
sns.lmplot(data=tips_df, x="total_bill", y="tip", hue="sex",markers=["*","^"])
```
